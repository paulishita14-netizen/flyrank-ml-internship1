## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

Unit of analysis: One row = one customer per month.
This means each observation is independent at the customer-month level.

A fixed rule like "if logins < 3 then churn = 1" fails because churn is driven by many interacting, non-linear factors.
The pattern is messy because:

1. Interaction effects: Low logins alone don't cause churn if the customer has high support tickets and a recent discount. ML can learn combinations.
2. Non-linear relationships: The risk of churn doesn't increase linearly with inactivity. The 1st missed month vs 3rd missed month have very different weights.
3. High-dimensional features: We have 20+ features like usage frequency, payment delays, feature adoption, NPS score, etc. Writing nested if-else for all combinations is impossible to maintain.
4. Concept drift: What causes churn changes over time. ML can be retrained on new data, but a fixed rule would go stale.

So ML is better here because it can learn complex, weighted patterns from data instead of relying on brittle human-defined thresholds.
